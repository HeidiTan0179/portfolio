```python
plt.hist(df.completion_year, ec='black')
plt.title('Completion Years of Properties')
plt.xlabel('Year')
plt.plot()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='completion_year', y='monthly_rent', data=df)
plt.title('Relationship Between Completion Year and Monthly Rent')
plt.xlabel('year')
plt.ylabel('monthly rent (ringgit)')
plt.show()     
```

```python
completion_monthly = df.groupby(['completion_year'])[['monthly_rent']].mean()
completion_monthly = completion_monthly.nlargest(10,'monthly_rent')
completion_monthly
```

```python
completion_monthly['monthly_rent'].plot(kind='bar', color='skyblue', ec='black')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.6)
plt.title('Top 10 Years with Highest Monthly Rent', fontsize=14)
plt.xlabel('Completion Years')
plt.ylabel('Monthly Rent (RM)')
plt.xticks(rotation=45)
plt.show()
```
